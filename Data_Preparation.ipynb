# Prepare the data for modeling by splitting into Train/validate and Test data

```python
#Read the CSV containing 629 features
import pandas as pd

df= pd.read_csv('quora-question-pairs/final_features.csv')
```

```python
df=df.drop('id', axis=1)
#Checking whether there are any rows with null values
nan_rows = df[df.isnull().any(1)]
df = df.dropna( how='any',
                    subset=['is_duplicate'])
```

```python
#Split the dataset into 80:20 using stratify
from sklearn.model_selection import train_test_split
x = df
y = df['is_duplicate']
x_train, x_test, y_train,y_test = train_test_split(x, y, test_size = 0.2,  stratify = y)
```

```python
df_train = pd.DataFrame(x_train)
df_test = pd.DataFrame(x_test)
```
